```python
#Loading the dependencies
import boto3
import pandas as pd
from io import StringIO
```

```python
#Loading the Csv's
noworld_population_df = pd.read_csv("Emission_Population_L5000_Data.csv") 
noworld_population_df.tail()
```

```python
emissions_N2O_new_df = noworld_population_df[noworld_population_df['Item_Code'].isin([5061, 6516, 6992, 6995, 6996]) & 
                                        noworld_population_df['Element_Code'].isin([7230, 7234, 7236]) &
                                        (noworld_population_df['Year'] > 2010 )&(noworld_population_df['Emission']>5) &
                                            (noworld_population_df['Population'] > 5000) ]


emissions_N2O_new_df
```

```python
filename = 'EmissionN2O.csv' #Any name we want to have in bucket
bucketname = 'bucketname' #S3 bucket name

csv_buffer = StringIO()
emissions_N2O_new_df.to_csv(csv_buffer)

## in AWS to go click ur name and then "My Security Credentials". Then "Access Keys"
client = boto3.client('s3', 
        region_name = 'region', #need to change here
        aws_access_key_id = 'Access-Key', #need to change here
        aws_secret_access_key = 'Secret-access-key' #need to change here
)

response = client.put_object(
    ACL='private',
    Body=csv_buffer.getvalue(),
    Bucket=bucketname,
    Key=filename
)
```
